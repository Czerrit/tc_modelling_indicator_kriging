```python
unique_entries = df['Facies'].unique()

# Ausgabe der einzigartigen EintrÃ¤ge
print(unique_entries)
```

```python
interval_size = 0.5 #m
big_new_df = pd.DataFrame()
groups = df.groupby(['id'])
for id, subdf in groups:
    subdf = subdf[::-1]
    new_df = pd.DataFrame()
    all_new_ys = []
    all_new_zs = []
    all_new_facies = []
    for i in range(len(subdf)-1):
        a = subdf['Z'].values[i]
        b = subdf['Z'].values[i+1]
        new_z = np.arange(b, a, -interval_size)[::-1]
        new_facies = list(np.repeat(subdf['Facies'].values[i], len(new_z)-1))
        new_facies.append(subdf['Facies'].values[i+1])
        new_y = np.repeat(subdf['Y'].values[i+1], len(new_z))
        for j in range(len(new_z)):
            all_new_zs.append(new_z[j])
            all_new_ys.append(new_y[j])
            all_new_facies.append(new_facies[j])

    new_df['Y'] = np.array(all_new_ys).ravel()[::-1]
    new_df['Z'] =  np.array(all_new_zs).ravel()[::-1]
    new_df['Facies'] =  np.array(all_new_facies).ravel()[::-1]
    big_new_df = pd.concat((big_new_df, new_df))
```

```python
for facies in big_new_df['Facies'].unique():
    if facies in df['Facies'].unique():
        pass
    else:
        print('----', facies)
```
